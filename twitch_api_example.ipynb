```python
import twitch
from twitch import TwitchClient
```

```python
from dotenv import load_dotenv
import os

load_dotenv()
```

```python
client = twitch.TwitchHelix(
    client_id=os.getenv("twitch_api_client_id"),
    client_secret=os.getenv("twitch_api_client_secret"),
    scopes=[twitch.constants.OAUTH_SCOPE_ANALYTICS_READ_EXTENSIONS]
)
```

```python
client.get_oauth()
```

```python
streams = client.get_streams(user_logins=["distortion2", "ArmadaUGS"])[:10]
streams
```

```python
from PIL import Image
import requests
from io import BytesIO

w, h = 1600, 900

for stream in streams:
    response = requests.get(stream["thumbnail_url"].format(width=w, height=h))
    img = Image.open(BytesIO(response.content))
    display(img)
```
