```python
from PIL import Image

import pytesseract
```

```python
import os

example_screenshots_dir = "tests/data/example_screenshots"
os.listdir(example_screenshots_dir)
```

```python
def show_and_process_image(image, **kwargs):
    display(image)
    print(pytesseract.image_to_string(image, **kwargs))
```

```python
show_and_process_image(
    Image.open(example_screenshots_dir + "/text.png"),
    **{"lang": "eng"},
)
```

```python
show_and_process_image(
    Image.open(example_screenshots_dir + "/distortion2_2.png")
)
```

```python
show_and_process_image(
    Image.open(example_screenshots_dir + "/distortion2_2_overlay.png")
)
```

```python
from PIL.ImageOps import invert


with Image.open(example_screenshots_dir + "/distortion2_2_overlay.png") as im:
    scale = 3
    w, h = im.width, im.height
    show_and_process_image(invert(im.resize(size=(w * scale, h * scale)).convert("L").point(lambda p: 255 if p > 50 else 0)))
```

```python
with Image.open(example_screenshots_dir + "/distortion2_2.png") as im:
    scale = 2
    w, h = im.width, im.height
    show_and_process_image(invert(im.resize(size=(w * scale, h * scale)).convert("L").point(lambda p: 255 if p > 50 else 0)))
```

```python
def show_and_process_image_2(image_path, th):
    with Image.open(image_path) as im:
        scale = 2
        w, h = im.width, im.height
        show_and_process_image(invert(im.resize(size=(w * scale, h * scale)).convert("L").point(lambda p: 255 if p > th else 0)))
```

```python
show_and_process_image_2(example_screenshots_dir + "/lunarjump_smw64.png", th=100)
```
