```python
from PIL import Image

import pytesseract
```

```python
import os

example_screenshots_dir = "tests/data/example_screenshots"
os.listdir(example_screenshots_dir)
```

```python
help(pytesseract.image_to_string)
```

```python
def show_and_process_image(image):
    display(image)
    print(pytesseract.image_to_string(image, lang="eng"))
```

```python
show_and_process_image(
    Image.open(example_screenshots_dir + "/text.png")
)
```

```python
show_and_process_image(
    Image.open(example_screenshots_dir + "/lunarjump_smw641.png")
)
```

```python
# try resizing and grayscale

with Image.open(example_screenshots_dir + "/lunarjump_smw641.png") as im:
    scale = 3
    w, h = im.width, im.height
    show_and_process_image(im.resize(size=(w * scale, h * scale)).convert("L").point(lambda p: 255 if p > 100 else 0))
```

```python
show_and_process_image(
    Image.open(example_screenshots_dir + "/distortion2_overlay.png")
)
```

```python
with Image.open(example_screenshots_dir + "/distortion2_overlay.png") as im:
    scale = 3
    w, h = im.width, im.height
    show_and_process_image(im.resize(size=(w * scale, h * scale)).convert("L").point(lambda p: 255 if p > 120 else 0))
```
